```
try:
  import cupy as cp
except:
  pass

import cv2
import matplotlib.pyplot as plt
import numpy as np
import time

from Models.CP_WOPT import CP_WOPT_Model
from Models.TT_WOPT import TT_WOPT_Model
from Models.TR_WOPT import TR_WOPT_Model
from metrics import RSE
from Utils import mask_img
```

```
# Read lena image
img_path = '/content/drive/My Drive/Tensor-Completion/Code/data/lena_red.bmp'

img = cv2.imread(img_path, 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (256, 256)) / 255
plt.imshow(img)
print("img shape: ", img.shape)

# If using cupy
img = cp.asarray(img)
```

```
# Mask image
missing_rate=0.5
img_missing = mask_img(img, missing_rate=missing_rate)
img_missing = np.nan_to_num(cp.asnumpy(img_missing))
plt.imshow(img_missing) # If using numpy
```

<h2>CP-WOPT</h2>

```
# Create and train model using gradient descent on CPU
model = CP_WOPT_Model(img_missing, 24, init="normal", optimization="gradient_descent", lr=1e-6, use_gpu=False)
start = time.time()
model.train(5000)
end = time.time()
print("CP-WOPT CPU time:", end-start)
```

```
# Create and train model using gradient descent on GPU
model = CP_WOPT_Model(cp.asarray(img_missing), 24, init="normal", optimization="gradient_descent", lr=1e-6, use_gpu=True)
start = time.time()
model.train(5000)
end = time.time()
print("CP-WOPT GPU time:", end-start)
```

<h2>TT-WOPT</h2>

```
# Create and train model using gradient descent on CPU
model = TT_WOPT_Model(img_missing, (1, 16, 16, 1), optimization="gradient_descent", lr=1e-6, use_gpu=False)
start = time.time()
model.train(5000)
end = time.time()
print("TT-WOPT CPU time:", end-start)
```

```
# Create and train model using gradient descent on GPU
model = TT_WOPT_Model(cp.asarray(img_missing), (1, 16, 16, 1), optimization="gradient_descent", lr=1e-6, use_gpu=True)
start = time.time()
model.train(5000)
end = time.time()
print("TT-WOPT CPU time:", end-start)
```

```
# Create and train model using gradient descent on CPU
model = TR_WOPT_Model(img_missing, (8, 8, 8, 8), optimization="gradient_descent", lr=1e-6, use_gpu=False)
start = time.time()
model.train(5000)
end = time.time()
print("TR-WOPT CPU time:", end-start)
```
